## Que es Markdown (.md)?

Colab tiene dos tipos de celdas: texto y código. Las celdas de texto se formatean utilizando un lenguaje de marcado simple llamado Markdown.


Para ver la fuente de Markdown, haga doble clic en una celda de texto, mostrando tanto la fuente de Markdown. Encima de la fuente de Markdown hay una barra de herramientas para ayudar en la edición.

## Diferencias entre Colab Markdown y otros dialectos de Markdown

Colab usa [marked.js](https://github.com/chjj/marked) y, por lo tanto, es similar, pero no del todo idéntico, al Markdown utilizado por Jupyter y Github.

Colab admite ecuaciones (MathJax) $\LaTeX$ como Jupyter, pero no permite etiquetas HTML en Markdown. Colab no admite algunas adiciones de GitHub, como emojis y casillas de tareas pendientes.

Si se debe incluir HTML en un cuaderno de Colab, consulte la [%%html magic](/notebooks/basic_features_overview.ipynb#scrollTo=qM4myQGfQboQ).
