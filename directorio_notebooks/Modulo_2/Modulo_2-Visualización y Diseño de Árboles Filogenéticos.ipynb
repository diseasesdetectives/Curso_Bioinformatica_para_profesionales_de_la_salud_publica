Microreact (https://microreact.org/) es una aplicación basada en web para proporcionar visualizaciones de datos interactivas. Permite la rápida generación y **vinculación de árboles**, **mapas**, **redes**, **gráficos** y **cronogramas**.





##Diseño a partir de Paneles

**Tipos de paneles**

Un panel dentro de Microreact contiene uno de varios tipos de datos que se pueden agregar y editar. Todos los paneles son configurables.

**Tabla de datos:** tabla estilo hoja de cálculo que muestra metadatos, que se puede utilizar para filtrar y dividir los datos que se muestran en otros paneles.

**Árbol:** visualización de un archivo de árbol filogenético

**Mapa:** datos como marcadores usando latitud/longitud de la tabla de datos, así como cualquier información geojson importada

**Línea de tiempo:** gráfico de una línea de tiempo específica para datos de fecha y hora seleccionados

**Red:** visualización de un archivo de datos de red.

**Gráficos:** gráficos predeterminados o totalmente personalizados que visualizan datos de la tabla de datos.
